```python
#importy
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#nacteni dat
data = pd.read_csv("dopravni_nehody.csv")
data.info()
```

```python
#prevedeni datum formatu a kontrola zacatku a konce sady
data['datum'] = pd.to_datetime(data['datum'])
print(data['datum'].max())
print(data['datum'].min())
```

```python
#filtrace dat pro minuly rok
start_date = pd.to_datetime('2023-01-01')
end_date = pd.to_datetime('2024-01-01')

data = data[(data['datum'] >= start_date) & (data['datum'] <= end_date)]
print(data['datum'].dtypes)

pd.options.display.max_info_columns = data.shape[1]
data.info()

print(data['datum'].max())
print(data['datum'].min())
```
